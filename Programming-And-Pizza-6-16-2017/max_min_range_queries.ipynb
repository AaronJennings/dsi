This workbook was the result of some data analysis during a programming and pizza workshop.  We were working on getting the row where a max and min value occurs when there are multiple rows with a measurement.

The ecology data carpentry data set was a good way to illustrate (https://figshare.com/articles/Portal_Project_Teaching_Database/1314459) get the surveys.csv file if you want to follow along.  

We were actually doing this in R with a different data set, but it's similar enough to python dataframes that it turned out to be transferable.  

```python
import pandas as pd

from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

surveys = pd.read_csv("data/surveys.csv")
```

```python
pysqldf("SELECT max(hindfoot_length) from surveys WHERE species_id = 'NL'")
```

```python
pysqldf("SELECT record_id, month, day, year, plot_id, species_id, sex, MAX(hindfoot_length) AS max_hindfoot_length, \
        MIN(weight) AS min_weight FROM surveys GROUP BY(species_id)")
```

```python
# how to get just one without aggregation, we did this as a spot check
pysqldf("select MAX(hindfoot_length) from surveys where species_id = 'BA'")
```

Next task, what if you'd like to set a range for each species, and get the results that show up within that range?

We created a sample data set, start_end.csv, for a couple of species, with a start and end range.

```python
start_end = pd.read_csv("data/start_end.csv")
```

```python
start_end
```

```python
# now you can join with the start_end table and use the range
# set for each species to filter

pysqldf("SELECT * from surveys join start_end on \
surveys.species_id = start_end.species_id \
WHERE hindfoot_length >= start_end.Start \
AND hindfoot_length <= start_end.End \
ORDER BY surveys.species_id")
```
