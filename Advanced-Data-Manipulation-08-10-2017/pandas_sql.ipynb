```python
pysqldf = lambda q: sqldf(q, globals())
```

```python
species_df = pd.read_csv("species.csv")
```

```python
pysqldf("SELECT * FROM surveys_ten JOIN species_df \
    ON surveys_ten.species_id = species_df.species_id")
```

```python
analysis_df = pysqldf("SELECT genus, COUNT(genus), AVG(weight), AVG(hindfoot_length) FROM species_df \
    INNER JOIN surveys_df \
    ON surveys_df.species_id = species_df.species_id \
    WHERE weight IS NOT NULL \
    AND weight > 5 \
    GROUP BY(genus) \
    HAVING AVG(weight) > 10")
```
